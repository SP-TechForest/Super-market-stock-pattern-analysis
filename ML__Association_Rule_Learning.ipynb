```python
transaction = []
for i in range(data.shape[0]):
  transaction.append([str(data.values[i,j]) for j in range(data.shape[1])])
```

```python
len(transaction)
```

```python
from apyori import apriori
rules = apriori(transactions= transaction, min_support = 0.003, min_confidence = 0.2, min_lift = 3, min_length = 2, max_length = 2)
```

```python
results = list(rules)
```

```python
def inspect_rules(results):
  for i in range(len(results)):
    lhs = [tuple(result[2][0][0])[0] for result in results]
    rhs = [tuple(result[2][0][1])[0] for result in results]
    confi = [result[2][0][2] for result in results]
    sup = [result[1] for result in results]
    lift = [result[2][0][3] for result in results]
    return list(zip(lhs, rhs, confi, sup, lift))
```

```python
report_rules = pd.DataFrame(inspect_rules(results), 
                            columns = ["Left_Hand_Side", "Right_Hand_Side", 
                                      "Confidence", "Support", "Lift"])
```
